```python
import geopandas as gpd
import matplotlib.pyplot as plt
import pandas as pd
```

```python
gdf = gpd.read_file('799_HMM_FZ.shp')
df = pd.read_csv('PinPoint 799 2021-10-08 10-04-41_activity.csv')
```

```python
df['MT Time'] = pd.to_datetime(df['MT Time'])
gdf['MT Time'] = pd.to_datetime(df['MT Time'])
```

```python
time_range = gdf['MT Time'].to_list()
gdf['ODBA_M'] = None
gdf['ODBA_STD'] = None
for x in time_range:
    time_1 = x  - pd.Timedelta(minutes = 7, seconds = 30)
    time_2 = x + pd.Timedelta(minutes = 7, seconds = 30)
    mask = (df['MT Time'] >= time_1) & (df['MT Time'] <= time_2)
    x_df = df[mask]
    gdf.loc[df['MT Time'] == x, 'ODBA_M'] = x_df['ODBA'].mean()
    gdf.loc[df['MT Time'] == x,'ODBA_STD'] = x_df['ODBA'].std()
```
