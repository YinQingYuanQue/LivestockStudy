```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as plt
from tqdm import tqdm
import numpy as np
%matplotlib inline
```

Link the Fuzzy table to GPS file

```python
GPS_939 = pd.read_csv('../Data/GPS/2022/CSV/939_cl_R.csv')
FZ_939 = pd.read_csv('../testFuzzy/Fuzzy_table/Output_939_2.csv', header = None)
df_939 = pd.merge(GPS_939, FZ_939, left_on='No', right_on= 3, how='inner')
df_939.head()
```

```python
df_939['time'] = pd.to_datetime(df_939['time'])
df_939.set_index('time', inplace=True)
df_939 = df_939.between_time('7:00:00','20:00:00')
df_939 = df_939.reset_index()

oct_df = df_939[df_939['time'].dt.month == 10]
nov_df = df_939[df_939['time'].dt.month == 11]
dec_df = df_939[df_939['time'].dt.month == 12]
jan_df = df_939[df_939['time'].dt.month == 1]
feb_df = df_939[df_939['time'].dt.month == 2]
mar_df = df_939[df_939['time'].dt.month == 3]
apr_df = df_939[df_939['time'].dt.month == 4]
may_df = df_939[df_939['time'].dt.month == 5]
jun_df = df_939[df_939['time'].dt.month == 6]
```

```python
combination = []
all_p = range(0,3**12)
for i2 in all_p:
    Tri = []
    Quotient = int(i2/3)
    Reminder = i2 % 3
    Tri.append(Reminder)
    while Quotient != 0:
        Reminder = Quotient % 3
        Tri.append(Reminder)
        Quotient = int(Quotient/3) 
    combination.append(Tri)
    #print (Tri)
#print (combination)

comb = []
for i in combination:
    j = i[:]
    l = len(j)
    while l != 16:
        j.append(0)
        l = l + 1
    comb.append(j)
    #print (j,j[0]*3**0 + j[1]*3**1 + j[2]*3**2 + j[3]*3**3 + j[4]*3**4 + j[5]*3**5
    #      + j[6]*3**6 + j[7]*3**7 + j[8]*3**8 + j[9]*3**9 + j[10]*3**10 + j[11]*3**11
    #      + j[12]*3**12 + j[13]*3**13 + j[14]*3**14)
#print (comb)
        
        
```

```python
tmp_df = oct_df[oct_df['time'].dt.day == 16]
tmp_df[-10:]
```

```python
cum_exp = []
for i in tqdm(range(1,32)):
    not_enough_data = 0
    hourly_list = []
    try:
        tmp_df = oct_df[oct_df['time'].dt.day == i]
    except:
        continue
    for j in range(8,20):
        if tmp_df[tmp_df['time'].dt.hour == j].shape[0] > 0: # enough records in current hour
            hourly_list.append([tmp_df[tmp_df['time'].dt.hour == j].iloc[0][0], 
                                tmp_df[tmp_df['time'].dt.hour == j].iloc[0][1], 
                                tmp_df[tmp_df['time'].dt.hour == j].iloc[0][2],
                                tmp_df[tmp_df['time'].dt.hour == j].iloc[0]['RASTERVALU']])
            
        elif tmp_df[tmp_df['time'].dt.hour == j - 1].shape[0] > 0: # enough records in previous hour
            hourly_list.append([tmp_df[tmp_df['time'].dt.hour == j - 1].iloc[0][0], 
                                tmp_df[tmp_df['time'].dt.hour == j - 1].iloc[0][1], 
                                tmp_df[tmp_df['time'].dt.hour == j - 1].iloc[0][2],
                                tmp_df[tmp_df['time'].dt.hour == j - 1].iloc[0]['RASTERVALU']])
            
        elif tmp_df[tmp_df['time'].dt.hour == j + 1].shape[0] > 0: # enough records in post hour
            hourly_list.append([tmp_df[tmp_df['time'].dt.hour == j + 1].iloc[0][0], 
                                tmp_df[tmp_df['time'].dt.hour == j + 1].iloc[0][1], 
                                tmp_df[tmp_df['time'].dt.hour == j + 1].iloc[0][2],
                                tmp_df[tmp_df['time'].dt.hour == j + 1].iloc[0]['RASTERVALU']])
            
        else:
            not_enough_data = 1
            break
    
    if not not_enough_data:
        df_hourly = pd.DataFrame(hourly_list, columns = ['FM_L', 'FM_M', 'FM_H', 'RASTERVALU'])
        C_r = []
        C_p = []
        for i3 in comb[0:3**12]:
            n = 0
            s = 0
            p = 1

            for j in range(0,12):
                if i3[j] == 0:
                    s = s +  df_hourly.loc[n,'RASTERVALU'] * 0.4
                    p = p * (df_hourly.loc[n,'FM_L']/(df_hourly.loc[n,'FM_L']+df_hourly.loc[n,'FM_M']+df_hourly.loc[n,'FM_H']))
                if i3[j] == 1:
                    s = s +  df_hourly.loc[n,'RASTERVALU'] * 0.2
                    p = p * (df_hourly.loc[n,'FM_L']/(df_hourly.loc[n,'FM_L']+df_hourly.loc[n,'FM_M']+df_hourly.loc[n,'FM_H']))
                if i3[j] == 2:
                    s = s +  df_hourly.loc[n,'RASTERVALU'] * 0.4
                    p = p * (df_hourly.loc[n,'FM_L']/(df_hourly.loc[n,'FM_L']+df_hourly.loc[n,'FM_M']+df_hourly.loc[n,'FM_H']))
                n = n + 1
        #    print 'i=',i,'s=',s,'p=',p
            C_r.append(s)
            C_p.append(p)

        C = pd.DataFrame(data = [C_r,C_p]).T
        C.columns = ['Cr','Cp']
        #C = C.sort_values(by=['Cr'])
        s = C['Cr']*C['Cp']

        ####
        #Monthly.append(s)
        #Monthly--->csv
        ####
        print (i,s.sum())
        cum_exp.append([i,s.sum()])
    else:
        print (i,'Nan')
        cum_exp.append([i,-1])
cum_exp = pd.DataFrame(cum_exp)
cum_exp.to_csv('../Analysis/Cum_Exp/Leo_939_Oct.csv')
```
